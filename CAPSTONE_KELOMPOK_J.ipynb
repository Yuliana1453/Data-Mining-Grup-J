## Data Understanding:

1. **Nomor Induk Mahasiswa (NIM)** = Nomor Induk Mahasiswa yang digunakan sebagai identifikasi unik alumni.
2. **Tahun Masuk UIN Jakarta** = Tahun alumni mulai menempuh pendidikan di UIN Jakarta.
3. **Tahun Wisuda** = Tahun alumni menyelesaikan sidang dan diwisuda. Bisa diisi dengan tahun sidang jika lebih relevan.
4. **Bulan Wisuda** = Bulan saat alumni diwisuda.
5. **No HP (WhatsApp)** = Nomor telepon alumni.
6. **email** = Email alumni.
7. **Waktu Mendapatkan Pekerjaan** = Lama waktu (dalam bulan) dari saat wisuda hingga alumni mendapatkan pekerjaan pertama.
8. **Status saat ini** = Status terkini alumni, apakah mereka sedang melamar pekerjaan, mengurus rumah tangga, bekerja di sektor BUMN, PNS, swasta, berwirausaha, melanjutkan studi, atau menjadi pengajar.
9. **Tahun Wirausaha** = Tahun alumni memulai usaha, khusus bagi alumni yang menjadi wirausaha.
10. **Bidang wirausaha** = Keterangan singkat tentang bidang usaha yang dijalankan oleh alumni, khusus bagi yang berwirausaha.
11. **Perusahaan** = Nama perusahaan atau instansi tempat alumni bekerja saat ini.
12. **Posisi** = Posisi atau jabatan alumni di tempat kerja mereka saat ini.
13. **Penghasilan** = Perkiraan penghasilan bulanan alumni (take-home pay) dalam juta rupiah.
14. **Univeritas** = Nama universitas dan kota tempat alumni melanjutkan studi, jika ada.
15. **Tahun Masuk Studi Lanjut** = Tahun ketika alumni melanjutkan studi setelah lulus dari UIN Jakarta.
16. **Nama jurusan/program studi setelah Kuliah di FST UIN Jakarta** = Nama jurusan atau program studi yang diambil alumni saat melanjutkan pendidikan.
17. **Kesesuaian Kurikulum** = Skor yang diberikan alumni mengenai kesesuaian kurikulum program studi mereka dengan kebutuhan lapangan pekerjaan. Skala 1-5.
18. **Kesesuaian Bidang Ilmu** = Skor yang diberikan alumni tentang seberapa sesuai bidang ilmu yang mereka pelajari dengan pekerjaan mereka saat ini. Skala 1-5.
19. **Saran** = Komentar, kesan, dan saran alumni untuk pengembangan program studi.
20. **Program Studi** = Program studi yang diambil oleh alumni saat menempuh pendidikan di UIN Jakarta. Ada beberapa pilihan seperti Agribisnis, MAG (Magister), Biologi, Fisika, Kimia, Matematika, SI (Sistem Informasi), TI (Teknik Informatika), Teknik Pertambangan, dan MTI (Magister Teknik Informatika).
21. **Lama Bekerja di Institusi Saat ini** =  Lama waktu alumni bekerja di institusi/perusahaan tempat mereka bekerja saat ini.
22. **Kota tempat bekerja saat Ini** = Kota di mana alumni bekerja saat ini.

## Membuat Variabel Baru

### Kategori Posisi

```python
# Fungsi untuk mengategorikan posisi
def categorize_position(pos):
    if pd.isna(pos):
        return "Tidak Diketahui"
    elif "admin" in pos.lower():
        return "Admin"
    elif "staff" in pos.lower():
        return "Staff"
    elif "engineer" in pos.lower():
        return "Engineer"
    elif "guru" in pos.lower() or "pengajar" in pos.lower():
        return "Pendidikan"
    elif "marketing" in pos.lower():
        return "Marketing"
    elif "qc" in pos.lower() or "quality control" in pos.lower():
        return "Quality Control"
    elif "data" in pos.lower() or "analyst" in pos.lower():
        return "Data Analyst"
    elif "developer" in pos.lower() or "programmer" in pos.lower():
        return "IT Developer"
    elif "manager" in pos.lower():
        return "Manager"
    else:
        return "Lainnya"

# Terapkan fungsi ke kolom 'Posisi' untuk membuat kolom 'Kategori_Posisi'
Data['Kategori Posisi'] = Data['Posisi'].apply(categorize_position)

# Dapatkan posisi kolom 'Posisi'
pos = Data.columns.get_loc('Posisi') + 1

# Sisipkan kolom 'Kategori Posisi' tepat setelah kolom 'Posisi'
Data.insert(pos, 'Kategori Posisi', Data.pop('Kategori Posisi'))
```

### Kategori Perusahaan

```python
# Fungsi untuk mengategorikan perusahaan
def categorize_company(Comp):
    if pd.isna(Comp):
        return "Tidak Diketahui"
    elif "bank" in Comp.lower() or 'asuransi' in Comp.lower():
        return "Keuangan"
    elif 'teknologi' in Comp.lower() or 'it' in Comp.lower() or 'digital' in Comp.lower():
        return 'Teknologi'
    elif 'sekolah' in Comp.lower() or 'universitas' in Comp.lower() or 'yayasan' in Comp.lower():
        return 'Pendidikan'
    elif 'pertanian' in Comp.lower() or 'gizi' in Comp.lower() or 'pangan' in Comp.lower():
        return 'Pertanian'
    elif 'energi' in Comp.lower() or 'mineral' in Comp.lower():
        return 'Energi'
    elif 'kesehatan' in Comp.lower() or 'klinik' in Comp.lower() or 'puskesmas' in Comp.lower():
        return 'Kesehatan'
    else:
        return 'Lainnya'

# Terapkan fungsi ke kolom 'Perusahaan' untuk membuat kolom 'Kategori_Perusahaan'
Data['Kategori Perusahaan'] = Data['Perusahaan'].apply(categorize_company)

# Dapatkan posisi kolom 'Perusahaan' dan simpan dalam variabel 'Comp'
Comp = Data.columns.get_loc('Perusahaan') + 1

# Sisipkan kolom 'Kategori_Perusahaan' tepat setelah kolom 'Perusahaan' menggunakan 'Comp'
Data.insert(Comp, 'Kategori Perusahaan', Data.pop('Kategori Perusahaan'))
```

### Kategori wirausaha

```python
def categorize_business(business_type):
    if 'pendidikan' in business_type.lower() or 'bimbingan belajar' in business_type.lower() or 'training' in business_type.lower():
        return 'Pendidikan'
    elif 'fashion' in business_type.lower() or 'textile' in business_type.lower() or 'beauty' in business_type.lower():
        return 'Fashion & Kecantikan'
    elif 'kuliner' in business_type.lower() or 'olshop' in business_type.lower():
        return 'Kuliner'
    elif 'pertanian' in business_type.lower() or 'peternakan' in business_type.lower():
        return 'Agribisnis'
    elif 'consultant' in business_type.lower() or 'konsultan' in business_type.lower():
        return 'Konsultan'
    elif 'ticketing' in business_type.lower():
        return 'Pariwisata'
    elif 'printing' in business_type.lower() or 'souvenir' in business_type.lower() or 'seni' in business_type.lower() or 'kerajinan' in business_type.lower():
        return 'Kerajinan & Seni'
    elif 'geofisika' in business_type.lower() or 'software' in business_type.lower():
        return 'Teknik & Eksplorasi'
    elif 'kos' in business_type.lower():
        return 'Properti'
    elif 'jasa' in business_type.lower():
        return 'Jasa'
    else:
        return 'Lainnya'

# Terapkan fungsi ke kolom 'Perusahaan' untuk membuat kolom 'Kategori_Perusahaan'
Data['Kategori Usaha'] = Data['Bidang wirausaha'].apply(categorize_business)

# Dapatkan posisi kolom 'Perusahaan' dan simpan dalam variabel 'Comp'
business_type = Data.columns.get_loc('Bidang wirausaha') + 1

# Sisipkan kolom 'Kategori_Perusahaan' tepat setelah kolom 'Perusahaan' menggunakan 'Comp'
Data.insert(business_type, 'Kategori Usaha', Data.pop('Kategori Usaha'))
```
